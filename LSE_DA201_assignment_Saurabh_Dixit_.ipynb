####### **Group** the vaccinated data *by 'Province/State'* and Aggregate **Sum** to find the **Total Vaccincated** people **by Province/State**



```python
#Group the data frame by Province/State and calculate Total Vaccinated by Province

df_agg_vaccinated = vac.groupby(by=['Province/State'])[['Vaccinated']]\
.sum()\
.sort_values(by='Vaccinated')\
.reset_index()

df_agg_vaccinated
```

```python
#Bar chart showing Number of Recovered cases by Province/State
df_agg_vaccinated.plot.barh(x='Province/State',)
plt.title("No. of Vaccinated by each Province / State", \
          fontdict={'fontsize': 18,\
        'fontweight': 'bold',\
        'color': 'Black'})
plt.legend(loc='center right', fontsize='large')
plt.ylabel("Province / State", fontsize='16', fontdict={'fontweight':'bold'})
plt.xlabel("Number * 10^6", fontsize='16', fontdict={'fontweight':'bold'})
```

###### Notes and Observations
- **Lowest number** of Vaccinated people are in **'Saint Helena, Ascension and Tristan da Cunha'** Province/State
- **Highest number** of Vaccinated people are in **'Gibraltar'**

####  Limitations/Pitfalls :
The Total vaccincated numbers may be Low/High based on the **total population** of that Province/State. 
As such, to determine the percentage vaccinated by State we should look at the population figures as well. This should be one of areas to further explore.
