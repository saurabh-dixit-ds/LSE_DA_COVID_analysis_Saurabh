```python
# Define function to plot moving averages
def plot_moving_average(series, window, plot_intervals=False, scale=1.96):
    
    # Create a rolling window to calculate the rolling mean using the series.rolling function
    rolling_mean = series.rolling(window=window).mean()
    
    # Declare the dimensions for the plot, plot name and plot the data consisting of the rolling mean from above. 
    plt.figure(figsize=(18,4))
    plt.title('Moving average\n window size = {}'.format(window))
    plt.plot(rolling_mean, 'g', label='Simple moving average trend')

    
    # Plot confidence intervals for smoothed values
    if plot_intervals:
        
        # Calculate the mean absolute square 
        mae = mean_absolute_error(series[window:], rolling_mean[window:])
        
        # Calculate the standard deviation using numpy's std function
        deviation = np.std(series[window:] - rolling_mean[window:])
        
        # Calculate the upper and lower bounds 
        lower_bound = rolling_mean - (mae + scale * deviation)
        upper_bound = rolling_mean + (mae + scale * deviation)
        
        # Name and style upper and lower bound lines and labels 
        plt.plot(upper_bound, 'r--', label='Upper bound / Lower bound')
        plt.plot(lower_bound, 'r--')

    # Plot the actual values for the entire timeframe.
    plt.plot(series[window:], label='Actual values')
    plt.grid(True)
```

```python
# Define function to calculate the mean absolute error
def mean_absolute_error(a, b): return abs(b - a)
```

#### Ans to Q1 : Quantitative and Qualitative Data for Business Predictions
- Quantitative data is numeric in nature. For example, number of online purchases, length of a car, average numbe of hours for a journey and so on
- Qualitative data is categorical / descriptive in nature. For example, color, city, type of the car, sentiments.

In Business predictions, **both quantitative and qualitative data is important**. 

**For example:**
- **Quantitative data** like number of covid cases might indicate high number of cases in one region, but **qualitative data** can provide more context in terms of demographics in the region like age group, community, accessibility to medical facilities, poverty etc. which will influence the how far the pandemic could impact the population in that region. 
- **Quantitative data** like number of housing projects might indicate high availability of the Houses in one region, but **qualitative data** can provide more context in terms of buying power of the customers, earning potential, seasonal preferences within that region, Home Loan process complexity which will influence the number of houses we expect to be sold in that region

#### Ans to Q2 : Continuous Improvement
- Continuous Improvement is critical to remain relevant in today's world.
- Situations, Data, Demographics change constantly. The predictions we do are based on the date and trends we have seen until that momen, but this can drastically change 
- Continuous improvement is essentail for being efficient as well. As technology evolves there are better and faster ways to achieve results. If we don't continuously improve we will fall behind as a company / organisation and may not remain relevant in the market.
- Covid change the entire Travel and Healthcare industry for ever. Companies that are innovating and continuously improving can quickly adapt to changing circumstances and thrive in the new conditions


#### Ans to Q3 : Data Ethics - Is everything covered ?
- Data Ethics is an evolving topic. It has a direct correlation with Trust.
- Every organization dealing with data should have a well documented strategy for managing the **three pillars of Data Governance**
  - **The Organization** - what are the key organizational goals, what is the mission and the business model. What is the solution we are providing
  - **The Context** - what industry is the organization working in and what are the rules and regulations in that industry
  - **The Data** - how is data collected, stored, transformed, distributed, analyzed and deleted.
