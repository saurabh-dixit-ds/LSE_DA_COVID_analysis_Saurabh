```python
# ---- Address the question WHAT ARE TOTAL VACCINATIONS FOR A PARTICULAR REGION ----
```

```python
# make a copy of master data frame
df_ukcovid_vacc_copy = df_ukcovid_vacc_master

#validate the shape is exactly same as the master data frame
df_ukcovid_vacc_copy.shape
```

```python
df_ukcovid_vacc_copy.head()
```

```python
#Check if any missing values in Vaccinated col
df_ukcovid_vacc_copy['Vaccinated'].isna().value_counts()
```

```python
#Check if any missing values in Country/Region col
df_ukcovid_vacc_copy['Country/Region'].isna().value_counts()
```

```python
#Check if any missing values in Sub-region Name col
df_ukcovid_vacc_copy['Sub-region Name'].isna().value_counts()
```

```python
#Group the data frame by region name and calculate Total Vaccinated by Region
df_ukcovid_vacc_copy.groupby("Country/Region")[['Vaccinated']].sum()
```

```python
#Group the data frame by region name and calculate Total Vaccinated by Sub-region Name
df_ukcovid_vacc_copy.groupby('Sub-region Name')[['Vaccinated']].sum()
```

```python
#Group the data frame by Province/State and calculate Total Vaccinated by Province/State
df_vacc_by_state = df_ukcovid_vacc_copy.groupby(by=['Province/State'])[['Vaccinated']]\
.sum()\
.sort_values(by='Vaccinated', ascending=False)

#Show Total vaccinated grouped-by State in descending order
df_vacc_by_state
```

```python
#Group the data frame by Province/State and view total first dose
df_first_second_doses_by_state = df_ukcovid_vacc_copy.groupby(by=['Province/State'])[['First Dose','Second Dose']]\
.sum()\
.sort_values(by=['First Dose'], ascending=False)

#Show Total First Doses and Second Doses grouped-by Province/State
df_first_second_doses_by_state
```

```python
# Determine area(s) with the largest number of people who have received a first dose but no second dose
df_first_second_doses_by_state['Diff_First_Second'] = \
df_first_second_doses_by_state['First Dose'] \
- df_first_second_doses_by_state['Second Dose']

#DIFFERENCE between First and Second Dose numbers and see in Descending order
df_first_second_doses_by_state.sort_values(by=['Diff_First_Second'], ascending=False)
```

```python
# Create a copy of the master dataframe for UK Covid cases
df_uk_covid_cases_copy = df_ukcovid_cases_master
df_uk_covid_cases_copy.shape
```

```python
# Sense check the data 
df_uk_covid_cases_copy.head()
```

```python
# Below code answers Which area has the greatest number of recoveries

df_recoveries_by_state = df_uk_covid_cases_copy.groupby(by=['Province/State'])\
[['Recovered']]\
.sum()\
.sort_values(by=['Recovered'], ascending=False)

# Show Total Recoveries by Province/State
df_recoveries_by_state
```

```python
# Group by State/Province and show the number of dates by the date
df_uk_covid_deaths_by_State = df_uk_covid_cases_copy.groupby(by=['Province/State'])\
[['Deaths']]\
.sum()\
.sort_values(by=['Deaths'], ascending=False)
df_uk_covid_deaths_by_State
```
